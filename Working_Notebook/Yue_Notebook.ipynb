```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.dummy import DummyClassifier, DummyRegressor
import numpy as np
```

```python
df = pd.read_csv('../data/bigml_59c28831336c6604c800002a.csv')
df
```

```python
df = df.drop(columns = ['phone number', 'area code'])
```

```python
df['churn'].value_counts()
```

```python
df.info()
```

```python
df.head()
```

```python
df['average_hourly_night_rate'] = df['total night minutes'] / df['total night charge']
```

```python
df['average_hourly_day_rate'] = df['total day minutes'] / df['total day charge']
```

```python
df['average_hourly_intl_rate'] = df['total intl minutes'] / df['total intl charge']
```

```python
df['average_hourly_eve_rate'] = df['total eve minutes'] / df['total eve charge']
```

```python
df
```

```python
df.info()
```

```python
df.dropna(axis=0, inplace=True)
```

```python
df.info()
```

```python
X = df.drop('churn', axis=1)
y = df['churn']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
X_train_cat = X_train[['state', 'international plan', 'voice mail plan']]
```

```python
X_train_cat
```

```python
ohe = OneHotEncoder(categories = 'auto', drop='first', sparse=False, handle_unknown="error")
```

```python
X_train_encoded = ohe.fit_transform(X_train_cat)
```

```python
ohe.categories_
```

```python
X_train_encoded.shape
```

```python
X_train_encoded
```

```python
X_train_encoded = pd.DataFrame(
                        X_train_encoded,
                        columns=ohe.get_feature_names(), 
                        index=X_train.index,
)

X_train_encoded
```

```python
X_train_encoded.columns = X_train_encoded.columns.str.replace('x0_', '')
X_train_encoded = X_train_encoded.rename(columns= {'x1_yes': 'International Plan', 'x2_yes': 'Voicemail Plan'})
```

```python
X_train_encoded
```

```python
X_test_cat =  X_test[['state', 'international plan', 'voice mail plan']]
```

```python
X_test_encoded = ohe.transform(X_test_cat)
X_test_encoded
```

```python
X_test_encoded = pd.DataFrame(
                        X_test_encoded,
                        columns=ohe.get_feature_names(), 
                        index=X_test.index,
)

X_test_encoded
```

```python
X_test_encoded.columns = X_test_encoded.columns.str.replace('x0_', '')
X_test_encoded = X_test_encoded.rename(columns= {'x1_yes': 'International Plan', 'x2_yes': 'Voicemail Plan'})
```

```python
X_test_encoded
```

## Training Data

```python
X_train.drop(columns=['international plan', 'voice mail plan'], inplace=True)
```

```python
training_data = pd.concat([X_train, X_train_encoded, y_train], axis=1)
# training_data
training_data.to_csv('training_data.csv')
```

## Testing Data

```python
X_test.drop(columns=['international plan', 'voice mail plan'], inplace=True)
```

```python
testing_data = pd.concat([X_test, X_test_encoded, y_test], axis=1)
# testing_data
testing_data.to_csv('testing_data.csv')
```
